```python
from transformers import pipeline
```

```python
qa_pipeline = pipeline("question-answering", model="distilbert-base-uncased-distilled-squad")
```

```python
context = """
Regular physical activity is one of the most important things you can do for your health.
It can help control your weight, reduce your risk of heart disease, and strengthen your bones and muscles.
Exercise improves your mental health and mood by releasing endorphins, which are natural mood lifters.
People who are physically active tend to sleep better and have more energy during the day.
Engaging in at least 150 minutes of moderate-intensity aerobic activity per week, such as brisk walking or cycling,
is recommended by health experts. Strength training exercises, performed at least twice a week, help maintain muscle
mass and flexibility, especially as we age. A combination of aerobic and strength activities provides the greatest
health benefits. Moreover, regular exercise supports the immune system, helps manage blood sugar levels, and lowers
the risk of chronic illnesses such as type 2 diabetes and certain cancers. Even small amounts of daily activity can
make a difference in the long run, promoting both physical and mental well-being. Staying consistent is key, and
finding activities you enjoy can make it easier to turn exercise into a lifelong habit.
"""
```

```python
questions = [
    "How many minutes of aerobic activity are recommended per week?",
    "What chemicals does exercise release to improve mood?",
    "What are two benefits of strength training?",
    "How does physical activity affect mental health?"
]
```

```python
for q in questions:
    result = qa_pipeline(question=q, context=context)
    print(f"Q: {q}")
    print(f"A: {result['answer']}\n")
```

```python
def answer_question(question):
    if question.strip() == "":
        return "Please enter a question."
    result = qa_pipeline(question=question, context=context)
    return result["answer"]
```

```python
import gradio as gr
interface = gr.Interface(
    fn=answer_question,
    inputs=gr.Textbox(label="Ask a Question"),
    outputs=gr.Textbox(label="Answer"),
    title="Mini Question Answering System",
    description="Ask any question based on the given Health paragraph."
)
interface.launch()
```
